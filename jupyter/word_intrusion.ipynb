# Word Intrusion Test

## Import des librairies

```python
from embeddingTopicEvaluatorLib.models.bertopic_wrapper import TopicModelEvaluatorBERTopic
from datasets import load_dataset
import json

from embeddingTopicEvaluatorLib.human_evaluation.word_intrusion import generate_tasks, save_tasks
from embeddingTopicEvaluatorLib.human_evaluation.word_intrusion import word_intrusion_score
```

## Chargement des données

```python
db = load_dataset("ag_news",split="train")
docs = list(db["text"])
```

## Entraînement d'un modèle de topics

```python
from embeddingTopicEvaluatorLib.config.config import settings
config = settings.BERTOPIC_CONFIG_HDBSCAN
```

```python
model = TopicModelEvaluatorBERTopic(config=config)
model.evaluate(docs)
```

## Génération des mots de test et export en JSON

```python
tasks = generate_tasks(model, n_words=5, top_distant_pct=0.3, top_active_pct=0.2, useEmbeddingModel=True, stopWords=["the", "in", "a", "on", "into", "out", "an", "to", "of"])
print(json.dumps(tasks, indent=4))
save_tasks(tasks, "../data_word_intrusion/input/BERTopic_v2.json")
```

## Évaluation Word Intrusion Test

```python
with open("../data_word_intrusion/output/annotations.json", encoding="utf-8") as f:
    annotations = json.load(f)

score = word_intrusion_score(annotations)
print(f"Model Precision : {score:.2%}")
```
