# Topic Mixing Test

## Import des librairies

```python
from embeddingTopicEvaluatorLib.models.bertopic_wrapper import TopicModelEvaluatorBERTopic
from datasets import load_dataset
import json

from embeddingTopicEvaluatorLib.human_evaluation.topic_mixing import *
```

## Chargement des données

```python
db = load_dataset("ag_news",split="train")
docs = list(db["text"])
```

## Entraînement d'un modèle de topics

```python
from embeddingTopicEvaluatorLib.config.config import settings
config = settings.BERTOPIC_CONFIG_HDBSCAN
```

```python
model = TopicModelEvaluatorBERTopic(config=config)
model.evaluate(docs)
```

## Génération des mots de test et export en JSON

```python
tasks = generate_tasks_mixed(model=model, n_words=10, useEmbeddingModel=True)
print(json.dumps(tasks, indent=4))
save_tasks(tasks, "../data_topic_mixing/input/BERTopic.json")
```

## Évaluation Word Intrusion Test

```python
with open("../data_topic_mixing/output/annotations.json", encoding="utf-8") as f:
    annotations = json.load(f)

score = topic_mixing_score_mixed(annotations)
print(f"Model Precision : {score:}")
```
